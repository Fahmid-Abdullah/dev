## Duplicates

```python
ambassadors = pd.Series([
    'France',
    'United Kingdom',
    'United Kingdom',
    'Italy',
    'Germany',
    'Germany',
    'Germany',
], index=[
    'Gerard Araud',
    'Kim Darroch',
    'Peter Westmacott',
    'Armando Varricchio',
    'Peter Wittig',
    'Peter Ammon',
    'Klaus Scharioth'
])
ambassadors
```

```python
ambassadors.duplicated() # Checks for rows with duplicates
```

```python
ambassadors.duplicated(keep='last') # Marks entries as True if their value (country) is duplicated, except for the last occurrence
```

```python
ambassadors.duplicated(keep=False) # Marks all entries as True if their value (country) appears more than once
```

```python
ambassadors.drop_duplicates() # Drops duplicates but keeps the first occurrence of each duplicated value
```

```python
ambassadors.drop_duplicates(keep='last') # Drops duplicates except last occurrence
```

```python
ambassadors.drop_duplicates(keep=False) # Removes all occurrences of duplicated values  only keeps unique values
```

## Duplicates in Dataframes

```python
players = pd.DataFrame({ # Name and Pos are columns
    'Name': [
        'Kobe Bryant',
        'LeBron James',
        'Kobe Bryant',
        'Carmelo Anthony',
        'Kobe Bryant',
    ],
    'Pos': [
        'SG',
        'SF',
        'SG',
        'SF',
        'SF'
    ]
})
players
```

```python
players.duplicated() # Check columns with duplicates
```

```python
players.duplicated(subset=['Name']) # Check specifically Name column with duplicates
```

```python
players.duplicated(subset=['Name'], keep='last')
```

```python
players.drop_duplicates()
```

## Text Handling

### Splitting Columns

```python
df = pd.DataFrame({
    'Data': [
        '1987_M_US _1',
        '1990?_M_UK_1',
        '1992_F_US_2',
        '1970?_M_   IT_1',
        '1985_F_I  T_2'
]})
df
```

```python
df['Data'].str.split('_') # Splits on _
```

```python
df['Data'].str.split('_', expand=True) # Splits then expands into a dataframe
```

```python
df = df['Data'].str.split('_', expand=True) # Saves
df
```

```python
df.columns = ['Year', 'Sex', 'Country', 'No Children']
df
```

```python
df['Year'] = df['Year'].str.replace('?', '') # Takes ? out of any year entries
df
```

```python
df['Country'] = df['Country'].str.replace(' ', "") # Takes ' ' out of any country entries
df
```

```python
df['Country'].str.strip()
```

## More Visualization

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 11)
```

```python
plt.figure(figsize=(12, 6))

plt.title('My Nice Plot')

plt.plot(x, x**2, label='X Squared')
plt.plot(x, -1 * (x**2), label='X Cubed')
```

## OOP Interface

```python
fig, ax = plt.subplots(figsize=(12, 6))
```

```python
ax.plot(x, (x**2), color='red', linewidth=2, marker='o', markersize=8, label='X^2')
ax.plot(x, -1 * (x**2), color='blue', linewidth=2, marker='o', markersize=8, label='-X^2')

ax.set_title('My Nice Plot')
ax.set_xlabel('X')
ax.set_ylabel('x Squared')
ax.legend()

fig
```
