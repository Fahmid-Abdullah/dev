##Modifying DataFrames

```python
langs = pd.Series(
    ['French', 'German', 'Italian'],
    index=['France', 'Germany', 'Italy'],
    name='Language'
)
```

```python
df['Language'] = langs
```

```python
df
```

```python
df['Language'] = 'English'
```

```python
df
```

```python
df.loc[df['Language'] == 'English', 'Language'] = 'French'
```

```python
df
```

```python
df.rename(
    columns={
        'HDI': 'Human Development Index',
        'Anual Popcorn Consumption': 'APC'
    }, index={
        'United States': 'USA',
        'United Kingdom': 'UK',
        'Argentina': 'AR'
    })
```

```python
df.drop(columns='Language', inplace=True)
```

```python
df = pd.concat([
    df,
    pd.DataFrame([{
        'Population': 3,
        'GDP': 5
    }], index=['China'])
])
```

```python
df
```
